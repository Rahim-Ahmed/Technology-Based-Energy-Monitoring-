## ডেটা বিজ্ঞান
</div>

ডেটা বিশ্লেষণ ও ভিজ্যুয়ালাইজ করার জন্য আপনি Colab-এ জনপ্রিয় Python লাইব্রেরির সবচেয়ে উপযুক্ত ব্যবহার করতে পারেন। নিচের কোড সেলে <strong>numpy</strong> ব্যবহার করে র‍্যান্ডম ডেটা জেনারেট করা হয়েছে এবং <strong>matplotlib</strong> লাইব্রেরির সাহায্য সেগুলি ভিজ্যুয়ালাইজ করা হয়েছে। কোড এডিট করতে, শুধু সেলে ক্লিক করে এডিট করা শুরু করুন।

```python
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.plot(x="Timestamp", y=["Voltage"], kind="bar",linewidth=10)

plt.xlabel('Timestamp')
plt.ylabel('Voltage')
```

```python
plt.plot(df['Timestamp'], df['Power'],marker='o',linewidth=3)
plt.xlabel('Timestamp')
plt.ylabel('Power')
plt.title('Total power consumption vs Time')
plt.show()
```
